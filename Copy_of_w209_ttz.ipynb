```python
import yfinance as yf
import altair as alt
import pandas as pd
import datetime as dt
import numpy as np
import pandas_ta as ta
```

```python
start = dt.datetime(2020,1,2)
end = dt.datetime.now()
```

```python
# crude oil = yf.Ticker("^CL=F")
oil = yf.download("CL=F", start, end)
oil
```

```python
# sp_500 = yf.Ticker("^GSPC")
sp_500 = yf.download("^GSPC", start, end)
```

```python
sp_500.columns = sp_500.columns.to_flat_index()
sp_500 = sp_500.reset_index()
```

```python
sp_trim = sp_500[['Date', 'Adj Close']]
sp_trim = sp_trim.rename(columns={"Date": "Date", "Adj Close": "sp_close"})
sp_trim
```

```python
chart = alt.Chart(sp_trim).mark_line().encode(
    alt.X('Date:T', title='Date'),
    alt.Y('sp_close:Q'),
).properties(
    title='SP500 Close Price from 2021.1 to now')

chart
```

```python
oil.columns = oil.columns.to_flat_index()
oil_usd = oil.reset_index()
```

```python
oil_trim = oil_usd[['Date', 'Adj Close']]
oil_trim = oil_trim.rename(columns={"Date": "Date", "Adj Close": "oil_close"})
oil_trim 
```

```python
sp_trim['oil_close'] = oil_trim['oil_close']
sp_trim
```

```python
chart = alt.Chart(sp_trim).mark_line().encode(
    alt.X('Date:T', title='Date'),
    alt.Y('oil_close:Q'),
).properties(
    title='Crude Oil Close Price since 2021.1 to now'
)

chart
```

```python
base = alt.Chart(sp_trim).encode(x='Date:T')

alt.layer(
    base.mark_line(color='blue').encode(y='sp_close:Q'),
    base.mark_line(color='orange').encode(y='oil_close:Q')
)
```

```python
sp_trim
```

```python
sp_trim['sp_close_sq'] = np.sqrt(sp_trim['sp_close'])
sp_trim['oil_close_sq'] = np.sqrt(sp_trim['oil_close'])

sp_trim['sp_close_log'] = np.log(sp_trim['sp_close'])
sp_trim['oil_close_log'] = np.log(sp_trim['oil_close'])
sp_trim
```

```python
base = alt.Chart(sp_trim).encode(x='Date:T')

alt.layer(
    base.mark_line(color='blue').encode(y='sp_close_sq:Q'),
    base.mark_line(color='orange').encode(y='oil_close_sq:Q')
).properties(width=600)
```

```python
base = alt.Chart(sp_trim).encode(x='Date:T')

alt.layer(
    base.mark_line(color='blue').encode(y='sp_close_log:Q'),
    base.mark_line(color='orange').encode(y='oil_close_log:Q')
).properties(width=600)
```

```python
base = alt.Chart(sp_trim).encode(x='Date:T')

alt.layer(
    base.mark_line(color='blue').encode(y='sp_close_sq:Q'),
    base.mark_line(color='green').encode(y='oil_close_sq:Q')
).properties(width=600, title='S&P500 price (blue) vs. Crude Oil Close Price (green) since 2021.1 to now')
```
